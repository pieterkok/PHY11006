## Lecture 4: Motion and forces in 2 and 3 dimensions

Now that we have the necessary tools to describe motion in two and three dimensions, we can write Newton's second law in vector form:
\begin{align}
 \F = m \a\, .
\end{align}
The acceleration $\a$ always points in the same direction as $\F$. In this lecture we will first explore Newton's first law in two and three dimensions, and see how this leads to the notion of force balance. Next we will introduce the forces between multiple bodies and the centre of mass. Finally, we will introduce the notion of momentum, and re-express Newton's second law in terms of changes in momentum.


### Force balance

Newton's first law states that there is no *net* force on a body at rest or in uniform linear motion. This means that there may be multiple forces acting on the body, but they all sum to zero. In one dimension, this is easy to understand:
\begin{align}
 F_1 + F_2 + \ldots + F_N = 0\, .
\end{align}
All these forces lie along the same line.
In three dimensions, this becomes a *vector* equation:
\begin{align}
 \F_1 + \F_2 + \ldots + \F_N = 0\, ,
\end{align}
which should be understood as 
\begin{align}
 \begin{pmatrix}
  F_{1,x} \\ F_{1,y} \\ F_{1,z} 
 \end{pmatrix} 
 + 
 \begin{pmatrix}
  F_{2,x} \\ F_{2,y} \\ F_{2,z} 
 \end{pmatrix} 
 + \ldots + 
 \begin{pmatrix}
  F_{N,x} \\ F_{N,y} \\ F_{N,z} 
 \end{pmatrix}
 =
 \begin{pmatrix}
  0 \\ 0 \\ 0 
 \end{pmatrix}
 \equiv 0\, ,
\end{align}
Therefore, *each component* should add up to zero:
\begin{align}
 F_{1,x} + F_{2,x} + \ldots + F_{N,x} & = 0 \cr 
 F_{1,y} + F_{2,y} + \ldots + F_{N,y} & = 0 \cr
 F_{1,z} + F_{2,z} + \ldots + F_{N,z} & = 0 \, .
\end{align}
When you draw your diagrams, you will need to decompose your forces into their components.



As an example, consider a mass $m$ on a frictionless ramp that is subject to a force $f$ preventing it from sliding down (see Fig. 5). Gravity produces a downward force $W$, the weight, and the ramp produces a normal force $N$ that prevents the mass from moving into the ramp. The normal force will always have exactly the magnitude to keep it on the ramp. We can choose our $x$ and $y$ coordinates any way we like, as long as they are perpendicular (orthogonal) to each other. In this problem it is convenient to take $x$ down the ramp, and $y$ perpendicular to the ramp. Since the mass is at rest, the three forces should balance, i.e., sum to zero:
\begin{align}
 \mathbf{f} + \mathbf{N} + \mathbf{W} = 0\, .
\end{align}
In vector form this is
\begin{align}
 \begin{pmatrix} -f \\ 0 \end{pmatrix}  + \begin{pmatrix} 0 \\ N \end{pmatrix} + \begin{pmatrix} W\sin\theta \\ -W\cos\theta \end{pmatrix} = \begin{pmatrix} 0 \\ 0 \end{pmatrix} = 0\, .
\end{align}
Setting the $x$ and $y$ components to zero leads to the two equations
\begin{align}
 -f+ W\sin\theta = 0 \qquad\text{and}\qquad N - W \cos\theta = 0\, .
\end{align}
Since $W = mg$, and we want to know the values of the magnitudes $f$ and $N$ in terms of $m$ and $\theta$, we can rearrange
\begin{align}
 f = mg\sin\theta  \qquad\text{and}\qquad N = mg \cos\theta\, .
\end{align}
The normal force $N$ is due to Newton's third law, which states that a force on one body (the ramp) due to another (the mass) produces a force equal in magnitude and opposite in direction on the first body (the mass) due to the other (the ramp). Note the subtlety here that the normal force on the mass due to the ramp is *not* along $\mathbf{W}$, so it is the component perpendicular to the ramp that determines $\mathbf{N}$.

 > #### Exercise 
 > Solve the problem in figure 5 using coordinates where the $x$ direction is horizontal and the $y$ direction is vertical.

Another important concept for vectors is the *point of engagement*, or the starting point of the arrow that represents the force. We treated the mass in figure 5 as a point mass, and all forces therefore engage at the centre of gravity. However, for extended systems this can be different. For example, while the weight force ($mg$) always has the centre of gravity as the point of engagement, the normal force has the boundary between the block and the ramp surface as the point of engagement. Any friction force also has the point of engagement lying in the contact surface between the block and the ramp. We will explore the importance of this when we consider torque in a future lecture.

### Newton's third law
In the example of the block on the ramp in the previous section, the two bodies exert a force on each other: the weight of the block is a force *on* the ramp, while the ramp exerts an equal and opposite force on the block, namely the normal force. This is a consequence of Newton's third law that forces *always* appear in pairs that are equal and opposite, but acting on different objects. Imagine that object 1 exerts a force $\F_{12}$ on object 2. The point of engagement of $\F_{12}$ is on object 2. The third law states that object 2 also exerts a force on object 1, $\F_{21}$ that is opposite and equal in magnitude to $\F_{12}$:
\begin{align}
 \F_{21} = -\F_{12}\, .
\end{align}
The two objects do not need to be touching, as is the case with the block and the ramp. The gravitational force on the Earth by the Sun is a vector $\F_{\rm g}$ acting on the Earth and pointing towards the Sun. Similarly, the Earth exerts an equal force on the Sun in opposite direction, $-\F_{\rm g}$. The acceleration of the Earth $\a_{\rm E}$ due to the Sun's gravity can be written as 
\begin{align}
 \a_{\rm E} = \frac{\F_{\rm g}}{m_{\rm E}}\, ,
\end{align}
where $m_{\rm E}$ is the mass of the Earth. Similarly, the acceleration of the Sun $\a_{\odot}$ due to the Sun's gravity can be written as 
\begin{align}
 \a_{\odot} = \frac{\F_{\rm g}}{m_{\odot}}\, ,
\end{align}
where $m_{\odot}$ is the mass of the Sun. Even though the forces are the same in magnitude, the acceleration of the Earth is much larger than that of the Sun due to the great difference in masses $m_{\rm E}$ and $m_{\odot}$.

Often we want to consider a system of $N$ masses $m_1$, $m_2$, $\ldots$ $m_N$. This can be a galaxy of stars, the solar system, an atomic nucleus with orbiting electrons, or a formation of flying drones, to name a few examples. There may be forces between these objects. If there is a force $\F_{jk}$ on mass $m_k$ due to mass $m_j$, then there is also a force $\F_{kj} = -\F_{jk}$ on mass $m_j$ due to mass $m_k$ due to Newton's third law. 

### Momentum

At this point we introduce the *momentum* $\p$ of a body with mass $m$ and velocity $\v$ as $\p = m\v$. Momentum is a very useful concept when we want to quantify `how much motion' there is in a body. It plays a more fundamental role dan acceleration in Newton's second law, and we can write instead
```{math}
:label: vfgnjehiury87t43
 \F = \frac{\d\p}{\d t}\, .
```
The change in motion of a body is equal to the force exerted on that body. Equation {eq}`vfgnjehiury87t43` is equivalent to $\F = m\a$, but *only* when the mass is constant. There are however situations in which this is not the case. 


>#### Example: *Rocket motion* 
>
>Consider a rocket travelling with (increasing) velocity $\v$ relative to some rest frame (e.g., the Earth). The rocket expels hot gas with a velocity $\v'$ relative to the same rest frame, such that the exhaust velocity relative to the rocket is $\v_e = \v'-\v$. Assume furthermore that $\v_e$ is constant. We are interested in the force $\F$ on the rocket by the exhaust gas, and the velocity of the rocket after some time $t$.
> 
> The momentum of the rocket at time $t$ is $\p(t) = m\v(t)$, and a short time interval $\d t$ later, the momentum has become
>```{math}
>:label: johihjkjjjippj
> \p(t+\d t) = (m - \d m)(\v+\d\v) +\d m \v'  \, ,
>```
>where $\d m$ is the (positive) mass of hot gas expelled by the rocket. The first term in the right-hand side is the new momentum of the rocket, while the second term, $\d m\v'$, is the momentum of the expelled hot gas. Expanding the product in equation {eq}`johihjkjjjippj`, we obtain
>\begin{align}
 \p(t+\d t) = m\v +m\d\v -\v\d m +\d m \v' = m\v + m\d\v +\v_e \d m \, ,
\end{align}
>where we dropped the term that is quadratic in the infinitesimals, $\d m \d\v$, since they vanish in the final result (we nearly always drop products or powers of infinitesimals, since the vanish in the limit). The change in momentum is 
>\begin{align}
 \d\p = \p(t+\d t) - \p(t) = m\d\v +\v_e \d m \, .
\end{align}
>Hence
>\begin{align}
 \frac{\d\p}{\d t} = m\frac{\d\v}{\d t} +\v_e \frac{\d m}{\d t} \, .
\end{align}
>If there is an external force $\F$ on the rocket, such as Earth's gravity, then we have 
>\begin{align}
 \F =  \frac{\d\p}{\d t} = m\frac{\d\v}{\d t} +\v_e \frac{\d m}{\d t} \, .
\end{align}
>If the rocket accelerates in free space without external forces, then 
>\begin{align}
 \F = 0 \quad\longrightarrow\quad \m\frac{\d\v}{\d t} +\v_e \frac{\d m}{\d t} = 0\, ,
\end{align}
>and the momentum imparted on the hot gas balances the momentum gained by the rocket and its remaining fuel.
>
> Let $\F = m\g$, the gravitational force close to the Earth's surface. Our equation of motion becomes
>\begin{align}
 \frac{\d\v}{\d t} + \frac{\v_e}{m} \frac{\d m}{\d t} = \g \, .
\end{align}
>Here $\v$ and $\v_e$ point upward, while $\g$ points downward. This then becomes a one-dimensional problem
>\begin{align}
 \frac{\d v}{\d t} + \frac{v_e}{m} \frac{\d m}{\d t} = -g\, ,
\end{align}
>where the positive direction is upward. Integrating this equation, we obtain
>\begin{align}
 \int_{v_0}^v \d v + v_e\int_{m_0}^m \frac{\d m'}{m'} = -g\int_0^t \d t\, ,
\end{align}
>or
>\begin{align}
 v = v_0 + v_e\ln \frac{m}{m_0} - gt \, .
\end{align}
>If $m$ is the final mass and $t$ is the time in which all fuel has been used up, then $v$ is the final speed of the rocket. Note that we have assumed that $g$ does not change while the rocket is burning its fuel, which is typically not a very good approximation for very large rockets that keep burning their fuel until well into the upper atmosphere. We also ignored air resistance.


### Centre of mass

The centre of mass position $\R$ of $N$ particles is the sum of all the positions of the particles, weighed by their masses:
\begin{align}%\label{eq:fnje90}
 \R = \frac{\sum_{j=1}^N m_j \r_j}{\sum_{j=1}^N m_j}\, .
\end{align}
Here, the $\sum$ symbol means that we sum over all $j$ from 1 to $N$. The positions of the particles with respect to the centre of mass are $\r_j' = \r_j - \R$, and if we use the $\r_j'$ for our description of the particles' motion, the centre of mass always lies at the origin. This is often very convenient in calculations.

The velocity of the centre of mass can be found directly by differentiating $\R$ with time:
\begin{align}
 \V = \frac{\d\R}{\d t} = \frac{1}{M} \sum_{j=1}^N m_j \frac{\d\r_j}{\d t} =  \frac{1}{M} \sum_{j=1}^N m_j \v_j = \frac{1}{M} \sum_j \p_j\, .
\end{align}
where $M=\sum_j m_j$ is the total, time independent, mass of the system. We can define the total momentum $\P = \sum_j \p_j$, such that 
\begin{align}
 \P = M\V\, .
\end{align}
This means that the total momentum is the total mass times the velocity of the centre of mass. When two constituent masses in this system exchange momentum, for example due to a collision, one part gains $\Delta\p$, while the other part loses the same amount, $-\Delta\p$. This is ensured by Newton's third law. That means that the overall momentum $\P$ remains the same, even if the constituent parts are a swirling mess of interacting bodies. Only an *external* force $\F_{\rm ext}$ can change $\P$:
\begin{align}
 \F_{\rm ext} = \frac{\d\P}{\d t}\, .
\end{align}
This is Newton's second law again. It means that we can solve the motion of complex bodies if we consider the position and momentum of the centre of mass. It is also why we can treat planets and stars as point objects.
