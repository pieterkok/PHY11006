## Lecture 3: Vectors

### Vectors in three dimensions

Any object in three dimensions has a position that is given by three numbers. In cartesian coordinates, these are $x(t)$, $y(t)$, and $z(t)$ with respect to some origin $(0,0,0)$. The position can be written as a time-dependent vector
\begin{align}
 \r(t) = 
 \begin{pmatrix}
  {x}(t) \\ {y}(t) \\ {z}(t) 
 \end{pmatrix} 
 \equiv 
 \begin{pmatrix}
  r_{x}(t) \\ r_{y}(t) \\ r_{z}(t) 
 \end{pmatrix}\, ,
\end{align}
and taking the time derivatives of each component yields the three-dimensional velocity and acceleration of the object: 
\begin{align}
 \v(t) = \frac{\d\r}{\d t} = 
 \begin{pmatrix}
  \dot{x}(t) \\ \dot{y}(t) \\ \dot{z}(t) 
 \end{pmatrix} 
 =
 \begin{pmatrix}
  v_{x}(t) \\ v_{y}(t) \\ v_{z}(t) 
 \end{pmatrix}
 \qquad\text{and}\qquad
  \a(t) = \frac{\d\v}{\d t} = 
 \begin{pmatrix}
  \dot{v}_x(t) \\ \dot{v}_y(t) \\ \dot{v}_z(t) 
 \end{pmatrix}
  =
 \begin{pmatrix}
  a_{x}(t) \\ a_{y}(t) \\ a_{z}(t) 
 \end{pmatrix}\, ,
\end{align}
where we use a subscript $x$, $y$, or $z$ to denote the velocity and acceleration components in that spatial direction. We also use a dot to denote time derivatives. 

Just like in the one-dimensional case, we can integrate velocity and acceleration with respect to time in order to find the position and velocity, respectively. We need to do this for each of the three components separately:
\begin{align}
 x(t) & = x(t_0) + \int_{t_0}^t v_x(t')\, \d t' \, , \cr 
 y(t) & = y(t_0) + \int_{t_0}^t v_y(t')\, \d t' \, , \cr
 z(t) & = z(t_0) + \int_{t_0}^t v_z(t')\, \d t' \, . 
\end{align}
We can write this as 
\begin{align}
 \r(t) = \r(t_0) + \int_{t_0}^t \v(t')\, \d t' \, . 
\end{align}
However, you should always remember that these are really three separate integrals. Similarly, we can write for the velocity
\begin{align}
 \v(t) = \v(t_0) + \int_{t_0}^t \a(t')\, \d t' \, . 
\end{align}
Another useful way in which we can write the position, velocity and acceleration in three dimensions is using the unit vectors $\basis{x}$, $\basis{y}$, and $\basis{z}$:
 \begin{align}
 \basis{x} = \begin{pmatrix} 1 \\ 0 \\ 0 \end{pmatrix} \, , \qquad
 \basis{y} = \begin{pmatrix} 0 \\ 1 \\ 0 \end{pmatrix} \, , \qquad
 \basis{z} = \begin{pmatrix} 0 \\ 0 \\ 1 \end{pmatrix} \, .
\end{align}
You have seen these before as $\basis{x} = \hat{\mathbf{i}}$, $\basis{y} = \hat{\mathbf{j}}$, and $\basis{z} = \hat{\mathbf{k}}$, but we will use this `new' notation that shows a more explicit link with the cartesian coordinates $x$, $y$, and $z$. This will be useful later when we employ other coordinate systems, like polar coordinates. With these unit vectors we can write the velocity and acceleration as 
 \begin{align}
 \v &= \frac{\d x}{\d t}\, \basis{x} + \frac{\d y}{\d t}\,\basis{y} + \frac{\d z}{\d t}\,\basis{z} \quad\text{and} \cr
 \a &= \frac{\d v_x}{\d t}\, \basis{x} + \frac{\d v_y}{\d t}\,\basis{y} + \frac{\d v_z}{\d t}\,\basis{z} = \frac{\d^2 x}{\d t^2}\, \basis{x} + \frac{\d^2 y}{\d t^2}\,\basis{y} + \frac{\d^2 z}{\d t^2}\,\basis{z}\, .
\end{align}


### Tangent vectors and normal vectors to a curve

Consider a curved path of a particle, shown in Fig. 3. At time $t$ the particle is at position $A$ with position vector $\r$, and at time $t'$ later the particle is at position $B$ with position vector $\r'$. Define $\Delta t = t' - t$ and $\Delta \r = \r' -\r$. The line segment along the curve is $\Delta s$, which is a little bit longer than $\Delta\r$ because it is not straight. The average velocity $\v_{\rm gem}$ over the segment $\Delta\r$ is then defined as  
 \begin{align}
 \v_{\rm gem} = \frac{\Delta\r}{\Delta t}\, .
\end{align}
In the limit where $\Delta t\to0$, this becomes the time derivative $\v = \dot{\r}$, as expected:
 \begin{align}
 \v = \lim_{\Delta t \to 0} \frac{\Delta\r}{\Delta t} =  \lim_{\Delta t \to 0} \frac{\r(t+\Delta t) - \r(t)}{\Delta t} = \frac{\d\r}{\d t}\, ,
\end{align} 
where we have written $\r' = \r(t+\Delta t)$.
However, we can also see from Fig. 3 that in this limit the velocity $\v$ is the *tangent vector* to the curve. This is true for any continuous and differentiable path of a particle: the velocity at a point along the path is the tangent vector to the path at that point.

where we used the product rule for derivatives. The unit vector $\basis{T}$ may change direction along the curve, so the derivative $\d\basis{T}/\d t$ is generally not zero. In order to say a little bit more about this time dependence, we make the drawing in Fig. 4.

```{figure} ./images/radiuscurvature.png 
---

---
The radius of curvature $\rho$.
```
 
At the point $A$, we define the tangent and normal unit vectors 
\begin{align}
 \basis{T} = \basis{x} \cos\phi + \basis{y} \sin\phi \qquad\text{and}\qquad \basis{N} = -\basis{x} \sin\phi + \basis{y} \cos\phi\, .
\end{align}
We can take the time derivative of $\basis{T}$:
\begin{align}
 \frac{\d\basis{T}}{\d t} & = \basis{x} \frac{\d\cos\phi}{\d t} + \basis{y} \frac{\d\sin\phi}{\d t} = \, - \basis{x} \sin\phi\frac{\d\phi}{\d t} + \basis{y} \cos\phi\frac{\d\phi}{\d t} \, .
\end{align}
This yields the very simple relationship
\begin{align}
 \frac{\d\basis{T}}{\d t} = \basis{N}\frac{\d\phi}{\d t}\, .
\end{align}
We can now relate the time derivative of $\phi$ to the *radius of curvature* $\rho$ in Fig. 4 as follows:
\begin{align}
 \frac{\d\phi}{\d t} =  \frac{\d\phi}{\d s}  \frac{\d s}{\d t} = v  \frac{\d\phi}{\d s}\, ,
\end{align}
where $s$ is again the segment of the curve from $A$ to $A'$. From the triangle $MAA'$ in the figure, you can see that $\d s = \rho \d\phi$, and therefore $\d\phi/\d s = 1/\rho$. Hence 
\begin{align}
 \frac{\d\phi}{\d t} = \frac{v}{\rho} \qquad\text{and}\qquad \frac{\d\basis{T}}{\d t} = \frac{v}{\rho} \basis{N}\, .
\end{align}
Substituting this into the expression for the acceleration in equation {eq}`gbnhiue9`, we obtain
\begin{align}
 \a = \frac{\d v}{\d t} \basis{T} + \frac{v^2}{\rho} \basis{N}\, .
\end{align}
The first term is the component of acceleration or deceleration in the direction of the motion (i.e., tangential), while the second term is the normal acceleration that changes the direction of travel of the particle. You can see that in the case of uniform circular motion, where the speed is constant but the direction is always changing, the acceleration reduces to the familiar $v^2/R$, where the radius of curvature $\rho$ is equal to the radius of the circle $R$. 

For arbitrary curved paths, the radius of curvature is constantly changing, which results in quite dramatic movements of the point $M$. How does $M$ move as the path changes from an upwardly bent trajectory to a downward trajectory? The radius of curvature $\rho$ is more a conceptual tool for understanding the tangential and normal components of $\a$ than a computational tool for solving problems.



%\subsection{Relative position and velocity}
