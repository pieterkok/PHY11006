## Polar coordinates and circular motion

### Polar coordinates

When an object moves in a circle in the $xy$-plane around the origin $O$, the $x$ and $y$ coordinates as a function of time are given by 
```{math}
:label: fnju28
 x(t) = a \cos\omega t \qquad\text{and}\qquad y(t) = a \sin\omega t\, ,
```
where in this case the object lies on the $x$-axis at time $t=0$, and the radius of the circle is $a$. The *angular* velocity of the particle is $\omega$, to which we will return in a moment. The $x$ and $y$ coordinates are components of a two-dimensional vector 
\begin{align}
 \r(t) = \begin{pmatrix} x(t) \\ y(t) \end{pmatrix}\, ,
\end{align}
as we encountered before. While the Cartesian coordinates $x$ and $y$ (and $z$ in three dimensions) are often very convenient, for circular motion there are other coordinates that are more convenient, namely *polar* coordinates $r$ and $\theta$. Here $r$ is the distance to the origin, and $\theta$ is the angle between the $x$-axis and the line from the origin to the object's position. The reason these polar coordinates are more convenient is that for the circular motion above the distance to the origin stays constant ($r=a$), so we should be able to reduce the two-coordinate problem ($x$ and $y$) to a single-coordinate problem ($\theta$ only).

```{figure} ./images/polar2.png 
---
name: fig:dfhgjkgfhtd
width: 400px
---
Polar coordinates against a grid of Cartesian coordinates.
```

The relation between the Cartesian coordinates and the polar coordinates is given by 
```{math}
:label: nfru489weridgn
 x = r \cos\theta \qquad\text{and}\qquad y = r \sin\theta\, .
```
Note that these are always true, and carry a completely different meaning than the expressions in equation {eq}`fnju28`, which look superficially similar. Equation {eq}`fnju28` describes the motion of a particle over time, while equation {eq}`nfru489weridgn` describes the relation between two coordinate systems.

We can invert equation {eq}`nfru489weridgn` to get $r$ and $\theta$ in terms of $x$ and $y$:
```{math}
:label: mnjfdehiru9w
 r = \sqrt{x^2+y^2} \qquad\text{and}\qquad \theta = \arctan\left(\frac{y}{x}\right)\, .
```
This means that $r$ runs from zero to infinity (we take the positive square root), and $\theta$ runs from zero to $2\pi$ (This is another difference between equation {eq}`fnju28` and equation {eq}`nfru489weridgn`: the mapping between coordinate systems has to make sure the plane is covered exactly once by each coordinate system, whereas $\omega t$ can take values much greater than $2\pi$ as it describes the trajectory of a particle going around in circles indefinitely.):
\begin{align}
 r \in [0,\infty) \qquad\text{and}\qquad \theta \in [0,2\pi)\, .
\end{align}
We can write the motion of the object in equation {eq}`fnju28` in terms of polar coordinates as
```{math}
:label: jiohugvycftdr6ft
 r(t) = a = \text{constant}  \qquad\text{and}\qquad \theta(t) =  \omega t\, .
```
This means, as noted before, that we have to worry only about $\theta$ as a variable with respect to time. More general motion will again require two variables $r$ and $\theta$, or $x$ and $y$. As a rule of thumb, if the motion looks more like linear motion (but not quite), Cartesian coordinates may be better, and if it is closer to circular motion polar coordinates are likely better.

Returning to our example, we have to consider the basis vectors of our coordinate systems. For the Cartesian system this is simple:
```{math}
:label: kmlgjnerhitu2380eiwjof
 \r(t) = \begin{pmatrix} a\cos\omega t \\ a \sin\omega t \end{pmatrix} = a\cos\omega t \begin{pmatrix} 1 \\ 0 \end{pmatrix} + a\sin\omega t\begin{pmatrix} 0 \\ 1 \end{pmatrix} = a\cos\omega t\; \basis{x} + a\sin\omega t\; \basis{y}\, ,
```
where we introduced the basis vectors 
\begin{align}
 \basis{x}= \begin{pmatrix} 1 \\ 0 \end{pmatrix} \qquad\text{and}\qquad \basis{y} = \begin{pmatrix} 0 \\ 1 \end{pmatrix} \, .
\end{align}
Wherever we are in the $xy$-plane, the basis vector $\basis{x}$ always points in the direction of the positive $x$-axis, and the basis vector $\basis{y}$ always points in the direction of the positive $y$-axis. However, the basis vectors $\basis{r}$ and $\basis{\theta}$ point in different directions at different locations in the plane. The vector $\basis{r}$ always points away from the origin, while $\basis{\theta}$ is orthogonal to $\basis{r}$ and lies tangential to the circle centred around the origin and intersecting the point $(r,\theta)$. It points in the counter-clockwise direction.

```{figure} ./images/polar.png 
---
name: fig:ndgjkehiuro
---
Changing basis vectors in polar coordinates.
```

If you want to know the basis vectors $\basis{r}$ and $\basis{\theta}$ in terms of $\basis{x}$ and $\basis{y}$, you have to use trigonometry. You can readily infer from figure 10 that this gives
```{math}
:label: vngjehuor9
 \basis{r} = \cos\theta\; \basis{x} + \sin\theta\; \basis{y} \qquad\text{and}\qquad \basis{\theta} = -\sin\theta\; \basis{x} + \cos\theta\; \basis{y}\, .
```
We will use these relations a lot in what follows. The changing directions of $\basis{r}$ and $\basis{\theta}$ become particularly important when we take derivatives, for example when we want to work out the velocity and acceleration of an object. We will show how this works in a moment. First, what does a vector $\r$ look like in polar coordinates? It is remarkably compact:
```{math}
:label: gnehu94woriufd
 \r = r\, \basis{r} = a\, \basis{r}\, ,
```
where the second equality is specifically for our example. You should convince yourself by drawing $\r$ and $\basis{r}$ that there is no $\basis{\theta}$ contribution in the expression of the general vector $\r$ in polar coordinates.


### Derivatives of polar coordinates

For the object in circular motion in equation {eq}`fnju28`, we want to know what is its velocity. In Cartesian coordinates this follows the usual procedure. We use the recipe from the third lecture and work out the time derivative of $\r$:
```{math}
:label: nvgehuit394woej
 \v(t) = \frac{\d \r}{\d t} = \frac{\d }{\d t} \begin{pmatrix} a\cos\omega t \\ a \sin\omega t \end{pmatrix} = \begin{pmatrix} -a\omega\sin\omega t \\ a\omega \cos\omega t \end{pmatrix} \, .
```
We have to take derivatives of sines and cosines, but we know how to do that. The velocity of the particle has a constant magnitude 
\begin{align}
 v &= |\v(t)| = \sqrt{(-a \omega\sin\omega t)^2 + (a\omega \cos\omega t)^2} = a\omega \sqrt{\sin^2\omega t + \cos^2\omega t} \cr & = a\omega\, .
\end{align}
This is how the *angular* velocity $\omega$ relates to the regular velocity $v$. The greater the radius $a$ of the circle, the larger the velocity $v$ must be to keep up with the same angular velocity $\omega$.  The direction of $\v$ changes over time (as you can clearly visualise for circular motion), and in terms of Cartesian coordinates we can write this as 
\begin{align}
 \v = -v\sin\omega t \; \basis{x} + v\cos\omega t\; \basis{y}\, ,
\end{align}
where $\basis{x}$ and $\basis{y}$ are again (still!) constant at every position, and therefore constant over time, as our object moves from point to point in the plane.
From equation {eq}`vngjehuor9` we can see that the velocity of the circular motion always points in the tangential $\theta$ direction:
\begin{align}
 \v = v\, \basis{\theta}\, .
\end{align}
This reflects the fact that the basis vector $\basis{\theta}$ changes over time, since the direction of $\v$ changes over time but the magnitude $v$ is constant. This is a special case of the more general result we found in lecture 3, where the velocity is always tangent to the path of a particle.

Next, let's consider the velocity of the object in polar coordinates.  We take the time derivative of $\r$ in equation {eq}`gnehu94woriufd` to get
```{math}
:label: vgrho34jis
 \v(t) = \frac{\d\r}{\d t} =  a \frac{\d \basis{r}}{\d t}\, ,
```
where we used that $a$ is a constant. Clearly, we need to know how to take the time derivative of the basis vector $\basis{r}$. It cannot be zero, since that would make the velocity zero, and that is obviously wrong. The time derivative needs to be expressed in terms of  $\basis{r}$ and $\basis{\theta}$ themselves (which we can always do, because $r$ and $\theta$ form a proper coordinate system for any vector). We can start with equation {eq}`vngjehuor9` and use the fact that $\basis{x}$ and $\basis{y}$ do not change and thus their time derivative is zero:
\begin{align}
  \frac{\d\basis{r}}{\d t} &=  \frac{\d}{\d t} \left( \cos\theta\; \basis{x} + \sin\theta\; \basis{y} \right)  = -\sin\theta \frac{\d\theta}{\d t} \; \basis{x} + \cos\theta  \frac{\d\theta}{\d t}\; \basis{y} \cr
  & =  \frac{\d\theta}{\d t} \left(-\sin\theta\; \basis{x} + \cos\theta\; \basis{y} \right) = \frac{\d\theta}{\d t}\; \basis{\theta}\, .
\end{align}
Similarly,
\begin{align}
  \frac{\d\basis{\theta}}{\d t} &=  \frac{\d}{\d t} \left( -\sin\theta\; \basis{x} + \cos\theta\; \basis{y}  \right) = -\cos\theta \frac{\d\theta}{\d t} \; \basis{x} - \sin\theta  \frac{\d\theta}{\d t}\; \basis{y} \cr
  & = - \frac{\d\theta}{\d t} \left(\cos\theta\; \basis{x} + \sin\theta\; \basis{y} \right) = - \frac{\d\theta}{\d t}\; \basis{r}\, .
\end{align}
We collect these two important results in the following equation:
>```{math}
>:label: nrijweofdfd
>  \frac{\d\basis{r}}{\d t} =  \frac{\d\theta}{\d t}\; \basis{\theta} \qquad\text{and}\qquad \frac{\d\basis{\theta}}{\d t} = - \frac{\d\theta}{\d t}\; \basis{r}\, .
>```
These are the same relations between $\hat{e}_T$ and $\hat{e}_N$ in the third lecture.
Remembering that $\theta(t) = \omega t$, and substituting equation {eq}`nrijweofdfd` in the expression for $\v$, we have
\begin{align}
 \v(t) & = a  \frac{\d\theta}{\d t}\, \basis{\theta} = a\omega\,  \basis{\theta}\, ,
\end{align}
as required. Notice how the actual calculation in equation {eq}`vgrho34jis` using equation {eq}`nrijweofdfd` and equation {eq}`jiohugvycftdr6ft` is computationally very simple—much simpler than the derivatives of the sines and cosines! Here it is again in a streamlined version:

> The polar coordinates of the object are given by 
> \begin{align}
  r = a \qquad\text{and}\qquad \theta = \omega t\, .
 \end{align}
> The velocity $\v$ of the object is 
> \begin{align}
  \v = \frac{\d\r}{\d t} = a \frac{\d\basis{r}}{\d t} = a \frac{\d\theta}{\d t}\, \basis{\theta} = a\omega\, \basis{\theta}\, .
 \end{align}
> That's it. From this we can read off immediately that the speed is $v = a\omega$.

This calculation is so much simpler because the coordinate system matches the shape of the motion. You may object that equation {eq}`nrijweofdfd` was hard to find, but that is a *general* result, which holds for any motion in polar coordinates. You may still rather do the familiar calculus with sines and cosines in this case, but when we have non-uniform acceleration around the circle the calculation in Cartesian coordinates quickly become intractable. For example, consider the case where the rotation speed increases linearly with time ($\theta = \frac12\alpha t^2$). Calculate the vector $\v(t)$ in Cartesian coordinates and in polar coordinates and see the difference. Another example is the Archimedean spiral, where the radius increases linearly with $\theta$.

%Another good practice problem is finding the general second derivatives of $\basis{r}$ and $\basis{\theta}$ with respect to time.

### Circular motion

For uniform rotations in *three* dimensions, where the angular velocity $\omega$ is constant, there is another convenient and often-used description of circular motion that involves the *cross product*. The cross product between two vectors $\a$ and $\b$ is written as 
\begin{align}
 \a\times\b = (a_y b_z - a_z b_y) \basis{x} + (a_z b_x - a_x b_z) \basis{y} + (a_x b_y - a_y b_x) \hat{\z}\, .
\end{align}
This can be written as 
\begin{align}
 \a\times\b = ab\sin\theta \, \hat{e}_N\, ,
\end{align}
where $\theta$ is the angle between $\a$ and $\b$, and $\hat{e}_N$ is the vector normal to the plane spanned by $\a$ and $\b$. This plane has an orientation: if the fingers on your right hand point in the direction as you rotate the vector $\a$ to vector $\b$, your thumb will point in the direction of $\hat{e}_N$.

Define the angular velocity $\omega$ as
\begin{align}
 \omega = \frac{\d \theta}{\d t}\, ,
\end{align}
as before. The speed along a circle with radius $R$ is then 
\begin{align}
 v = \omega R\, .
\end{align}
However, in three dimensions not all possible points lie in the $xy$-plane, and we want a relation between $\v$, $\r$, and $\omega$. To this end we orient $\omega$ along the $z$-axis, so we can write $\vec{\omega} = \omega \hat{\z}$, and we pick a point $A$ along a circle with radius $R$ above the $xy$-plane. The position vector pointing to $A$ is $\r$. This is shown in Fig. 11.

```{figure} ./images/cirkelbeweging.png 
---
name: fig:lkghture4wjiohu
width: 250px
---
Vector relation between (constant) angular velocity $\vec{\omega}$, linear velocity $\v$ and position $\r$.
```

From the figure, $R = r\sin\gamma$, with $\gamma$ the angle between the vector $\hat{\z}$ and $\r$. Hence, the linear speed around the circle is
\begin{align}
 v = \omega r \sin\gamma\, .
\end{align}
However, this can be written as a cross product between $\vec{\omega}$ and $\r$:
\begin{align}%\label{eq:njbgtuehoi4wrnd}
 \v = \vec{\omega}\times\r\, .
\end{align}
This is true only for circular motion with constant $r$ and $\gamma$. While this is neat and compact, it is not so clear yet why this is a nice form. However, it allows us to treat the time derivative $\d/\d t$ in the case of constant circular motion as the cross product with the angular velocity: $\vec{\omega}\times\cdot$. We can find the acceleration from the time derivative of $\v$, and therefore for circular motion with constant $\omega$ we obtain
\begin{align}
 \a = \frac{\d\v}{\d t} = \frac{\d}{\d t} \vec{\omega}\times\r = \vec{\omega}\times \frac{\d\r}{\d t} = \vec{\omega}\times\v\, .
\end{align}
In its most useful form, we have 
\begin{align}%\label{eq:nfvjrowu32wireh}
 \a = \vec{\omega}\times(\vec{\omega}\times\r)\, .
\end{align}

Spherical coordinates are often the coordinates of choice when the potential depends only on $r$, the distance to the origin. They are defined by the relations
\begin{align}
 x & = r \cos\phi\sin\theta \cr
 y & = r \sin\phi\sin\theta \cr
 z & = r\cos\theta\, .
\end{align}
You should construct the inverse relation, specifying $r$, $\theta$, and $\phi$ in terms of $x$, $y$, and $z$.
